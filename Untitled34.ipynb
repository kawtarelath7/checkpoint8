```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd 
url = "https://en.wikipedia.org/wiki/Ben_Wellington_(rugby_league)"
R = requests.get(url)
R.status_code
soup = BeautifulSoup(R.content, "html")
soup.select_one("#firstHeading").text
[x.text for x in soup.select("#")]
```
